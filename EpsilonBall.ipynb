```python
epsilonballs = []
for file in os.listdir(path):
    if 'network_graph_POI' in file:
        print(file)
        G = pickle.load(open(path + '/' + file, "rb" ))
        edges = ox.graph_to_gdfs(G, nodes=False)
        coord = pd.DataFrame(index=['x','y'])
        for edge in edges.index:
            split = str(edges.iloc[edge].geometry).replace("LINESTRING (", '').replace(")",'').split(', ')
            for i in range(len(split)):
                x = split[i].split(' ')[0]
                y = split[i].split(' ')[1]
                coord = pd.concat([coord, pd.Series([x,y],index=['x', 'y'])],axis=1)
        coord = coord.T

        #epsilonballs.append(file)
        radius = 0.25#km
        a = (float(max(coord['x'])) - float(min(coord['x'])))/61
        x_li = []
        for i in range(60):
            x_li.append(float(min(coord['x']))+(i+1)*a)
        b = (float(max(coord['y'])) - float(min(coord['y'])))/61
        y_li = []
        for i in range(60):
            y_li.append(float(min(coord['y']))+(i+1)*b)
        d = 0
        #Für die einzelnen Epsilon Balls
        #print(float(min(coord['x'])),float(max(coord['x'])))
        for i in range(len(x_li)):
            #print(x_li[i])
            for j in range(len(y_li)):
                df = coord[abs(coord['y'].astype(float)-y_li[j]) <= abs(2*b)]
                df = df[abs(df['x'].astype(float)-x_li[i]) <= abs(2*a)]
                #sublist, in der nur werte sind, die bzgl. x und y werte im quadrat sind 
                n = 0
                for k in range(len(df)):
                    #dist = np.sqrt((float(coord.iloc[k]['x'])-x_li[i])**2 + (float(coord.iloc[k]['y'])-y_li[j])**2)
                    R = 6373.0
                    
                    lat1 = radians(float(df.iloc[k]['y']))
                    lon1 = radians(float(df.iloc[k]['x']))
                    lat2 = radians(float(y_li[j]))
                    lon2 = radians(float(x_li[i]))

                    dlon = lon2 - lon1
                    dlat = lat2 - lat1

                    e = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
                    c = 2 * atan2(sqrt(e), sqrt(1 - e))

                    distance = R * c
                    
                    if distance < radius:
                        n = 1
                        break
                if n == 0:
                    d += 1
        print(d)
        epsilonballs.append(d)
```
